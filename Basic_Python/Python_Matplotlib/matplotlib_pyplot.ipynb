````python
import matplotlib.pylab as plt
import numpy as np
````

# Plotting x and y points
The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

If we need to plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.

````python
x = np.array([1, 3, 5, 6, 7, 8])
y = np.array([1, 3, 5, 3, 2, 8])

plt.xlabel("X")
plt.ylabel("Y")

plt.plot(x, y)
plt.show()
````

# Plotting Without Line
To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.

````python
plt.plot(x, y, "X")
plt.show()
````

# Default X-Points
If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.

So, if we take the same example as above, and leave out the x-points, the diagram will look like this:

````python
plt.plot(y, marker = 'o')
plt.show()
````

# Format Strings `fmt`
You can also use the shortcut string notation parameter to specify the marker.

This parameter is also called `fmt`, and is written with this syntax:

```py
marker|line|color
```
## Line Reference
|Line Syntax|Description|
|-|-|
|'-'|Solid line|
|':'|Dotted line|
|'--'|Dashed line|
|'-.'|Dashed/dotted line|


```note
Note: If you leave out the line value in the fmt parameter, no line will be plotted.
```

````python
plt.plot(x, y, 'o--m') #marker|line|color
plt.show()
````

# Marker Size
You can use the keyword argument `markersize` or the shorter version, `ms` to set the size of the markers

````python
from turtle import color


plt.plot(x, y, marker='o', ms = 20, color='red') # also use can `color`
plt.show()
````

# Marker Color
You can use the keyword argument `markeredgecolor` or the shorter `mec` to set the color of the edge of the markers

````python
plt.plot(x, y, marker='o', ms='10', mec='red')
plt.show()
````

You can use the keyword argument `markerfacecolor` or the shorter `mfc` to set the color inside the edge of the markers

````python
plt.plot(x, y, marker='o', ms=10, mfc = 'r')
plt.show()
````

You can also use [Hexadecimal color values](https://www.w3schools.com/colors/colors_hexadecimal.asp)

````python
plt.plot(x, y, marker="o", ms=10, mec="#ff0000", mfc="#948fee")
plt.show()
````

# Linestyle
You can use the keyword argument `linestyle`, or shorter `ls`, to change the style of the plotted line

````python
plt.plot(x, y, linestyle = 'dotted')
plt.show()
````

### Shorter Syntax
The line style can be written in a shorter syntax:

`linestyle` can be written as `ls`.

`dotted` can be written as `:`.

`dashed` can be written as `--`.

````python
x1 = np.array([1, 3, 6, 8, 9, 10, 11])
y1 = np.array([3, 2, 1, 4, 1, 3, 5])

plt.plot(x, y, ls="dotted")
plt.plot(x1, y1, ls="-.")

plt.show()
````

# Line Color
You can use the keyword argument `color` or the shorter `c` to set the color of the line

````python
plt.plot(x1, y1, c="r")
plt.show()
````

# Line Width
You can use the keyword argument `linewidth` or the shorter `lw` to change the width of the line. The value is a floating number, in points.

````python
plt.plot(x1, y1, lw=4)
plt.show()
````

# Multiple Lines
You can plot as many lines as you like by simply adding more `plt.plot()` functions:

````python
plt.plot(x, y)
plt.plot(x1, y1)

plt.show()
````

We can also plot many lines by adding the points for the x- and y-axis for each line in the same `plt.plot()` function.

(In the examples above we only specified the points on the y-axis, meaning that the points on the x-axis got the the default values (0, 1, 2, 3).)

The x- and y- values come in pairs.

````python
plt.plot(x,y,x1,y1)
plt.show()
````
